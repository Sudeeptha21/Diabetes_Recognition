TO PREDICT DIABETES


```python
import pandas as pd
import numpy as np
import math
from sklearn.model_selection import train_test_split      #to split the data
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier 

```

```python
#read dataset
dataset = pd.read_csv('diabetes.csv')
dataset
```

```python
zero = ['Glucose','BloodPressure','SkinThickness','Insulin']
for column in zero:
    dataset[column]=dataset[column].replace(0,np.NaN)
    mean=int(dataset[column].mean(skipna=True))
    dataset[column]=dataset[column].replace(np.NaN,mean)
    print(dataset)
```

```python
#split dataset
x=dataset.iloc[:,0:8]
y=dataset.iloc[:,8]
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.3)
```

```python
#scaling data
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
#isolating y_train
y_t =[]
for i in y_train:
    y_t.append(i)
y_t
```

```python
#KNN
#finding k 
k=int(math.sqrt(len(y_test)))
print(k)
```

```python
#Euclidean distance
y_predicted =[]  #comps right answer
for point1 in x_test:
    distance = []
    for i,point2 in enumerate(x_train):
        d=[np.linalg.norm(point1-point2),y_t[i]]  #euclidean distance
        distance.append(d)
       
    distance.sort(key = lambda distance:distance[0])  #sorting column 1 i.e. distance 

    c=0
    for i in range(0,15):
        if distance[i][1] == 0:
            c=c+1
    if c>7:
        y_predicted.append(0)
    else:
        y_predicted.append(1)
y_predicted
    
```

```python
j=0
a1=0
a2=len(y_test)
for i in y_test:
    if i == y_predicted[j]:
        a1+=1
    j+=1
print(a1/a2)
```
